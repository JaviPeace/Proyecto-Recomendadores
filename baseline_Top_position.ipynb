```python
#Skill Columns

#Function to convert skill rating at each position.
def skillConverter(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    else:
        return val

skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for col in skill_columns:
    fifa[col] = fifa[col].apply(skillConverter)
    fifa[col].fillna(0.0, inplace = True)
```

```python
player_features = ['Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
       'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
       'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
       'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
       'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
df_postion  = pd.DataFrame()
for position_name, features in fifa.groupby(fifa['Position'])[player_features].mean().iterrows():
    top_features = dict(features.nlargest(5))
    df_postion[position_name] = tuple(top_features)
df_postion.head()
```

```python
position = []
player = []
club_l = []
for col in df_postion.columns:
    tmp_df = pd.DataFrame()
    l = [df_postion[col].values]
    l = l[0]
    l = list(l)
    l.append('Name')
    tmp_df = pd.DataFrame.copy(fifa[fifa['Position'] == col][l])
    tmp_df['mean'] = np.mean(tmp_df.iloc[: , :-1] , axis = 1)
    name = tmp_df['Name'][tmp_df['mean'] == tmp_df['mean'].max()].values[0]
    club = fifa['Club'][fifa['Name'] == str(name)].values[0]
    position.append(col)
    player.append(name)
    club_l.append(club)

gk = ['GK']
forward = ['LS', 'ST', 'RS','LF', 'CF', 'RF']
midfeilder = ['LW','RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM',
              'RCM', 'RM', 'LDM', 'CDM', 'RDM' ]
defenders = ['LWB','RWB', 'LB', 'LCB', 'CB','RB','RCB']

print('GoalKeeper : ')
for p , n , c in zip(position , player , club_l):
    if p in gk:
        print('{} [Club : {} , Position : {}]'.format(n , c , p))
print('\nFORWARD : ')
for p , n , c in zip(position , player , club_l):
    if p in forward:
        print('{} [Club : {} , Position : {}]'.format(n , c , p))
print('\nMIDFEILDER : ')
for p , n , c in zip(position , player , club_l):
    if p in midfeilder:
        print('{} [Club : {} , Position : {}]'.format(n , c , p))
print('\nDEFENDER : ')
for p , n , c in zip(position , player , club_l):
    if p in defenders:
        print('{} [Club : {} , Position : {}]'.format(n , c , p))
```

```python
from sklearn.metrics import ndcg_score
import numpy as np

def evaluate_by_club(dream_team_list, fifa, top_clubs, value_col='Value'):
    dream_team_sorted = sorted(
        dream_team_list,
        key=lambda name: fifa.loc[fifa['Name'] == name, value_col].values[0]
        if not fifa.loc[fifa['Name'] == name, value_col].empty else 0,
        reverse=True
    )

    dream_set = set(dream_team_list)
    results = {}
    # Mapear valor del jugador para los del dream team
    player_scores = {
        name: fifa.loc[fifa['Name'] == name, value_col].values[0]
        if not fifa.loc[fifa['Name'] == name, value_col].empty else 0
        for name in dream_team_sorted
    }


    for club in top_clubs:
        # Obtener jugadores del club (ground truth)
        gt_players = fifa[fifa['Club'] == club]['Name'].dropna().unique().tolist()

        # Ordenar ground truth por valor
        gt_sorted = sorted(
            gt_players,
            key=lambda name: fifa.loc[fifa['Name'] == name, value_col].values[0]
            if not fifa.loc[fifa['Name'] == name, value_col].empty else 0,
            reverse=True
        )

        y_true = [1 if player in dream_team_sorted else 0 for player in gt_sorted]


        prediction_ranking = [player for player in dream_team_sorted if player in gt_sorted]
        prediction_ranking += [player for player in gt_sorted if player not in prediction_ranking]

        y_score = [player_scores.get(player, 0) for player in gt_sorted]

        true_positives = set(dream_team_sorted) & set(gt_sorted)
        false_positives = set(dream_team_sorted) - set(gt_sorted)
        false_negatives = set(gt_sorted) - set(dream_team_sorted)

        precision = len(true_positives) / len(dream_team_sorted) if dream_team_sorted else 0
        recall = len(true_positives) / len(gt_sorted) if gt_sorted else 0
        novelty = 1 - recall if dream_team_sorted else 0

        ndcg = ndcg_score([y_true], [y_score]) if len(gt_sorted) > 0 else 0

        results[club] = {
            'precision': round(precision, 3),
            'recall': round(recall, 3),
            'novelty': round(novelty, 3),
            'ndcg': round(ndcg, 3),
            'true_positives': list(true_positives),
            'false_positives': list(false_positives),
            'false_negatives': list(false_negatives)
        }

    return results
```

```python
club_metrics = evaluate_by_club(dream_team_list, fifa, top_clubs, value_col='Value')

for club, metrics in club_metrics.items():
    print(f"\nResultados para {club}:")
    for key, val in metrics.items():
        if isinstance(val, list):
            continue
        print(f"  {key.capitalize()}: {val}")
```
